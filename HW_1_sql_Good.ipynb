2. Выберите топ 3 страны по количеству заказчиков, выведите их названия и количество записей.

```python
tasks+=1
pd.read_sql(
    '''
    SELECT Country AS 'Страна', COUNT(CustomerName) AS customer_count
    FROM Customers
    GROUP BY Country
    ORDER BY customer_count DESC
    LIMIT 3
    ''',
    con,
)
```

3. Выберите перевозчика, который отправил 10-й по времени заказ, выведите его название, и дату отправления.

```python
tasks+=1
pd.read_sql(
    '''
    SELECT ShipperName, OrderDate FROM Orders
    INNER JOIN Shippers ON Orders.ShipperID = Shippers.ShipperID
    ORDER BY substr(OrderDate,7, 4) || substr(OrderDate, 4, 2) || substr(OrderDate, 1, 2) ASC, OrderID ASC
    LIMIT 1 OFFSET 9
    ''',
    con,
)
#, OrderID - нужен в селекте для проверки (убрал)
```

4. Выберите самый дорогой заказ, выведите список товаров с их ценами.

```python
tasks+=1
pd.read_sql(
    '''
SELECT ProductName, Price
    FROM OrderDetails INNER JOIN Products ON OrderDetails.ProductID = Products.ProductID
    INNER JOIN(SELECT OrderID, SUM(position_price) AS order_price
    FROM (SELECT OrderID, Quantity * Price AS position_price
    FROM OrderDetails INNER JOIN Products ON OrderDetails.ProductID = Products.ProductID)
    GROUP BY OrderID ORDER BY order_price DESC LIMIT 1) AS TopPriceOrder
    ON OrderDetails.OrderID = TopPriceOrder.OrderID
    ''',
    con,
)
```

5. Какой товар больше всего заказывали по количеству единиц товара, выведите его название и количество единиц в каждом из заказов.

```python
tasks+=1
pd.read_sql(
    '''
    SELECT OrderDetails.OrderID, ProductName, OrderDetails.Quantity
    FROM Products INNER JOIN OrderDetails ON Products.ProductID = OrderDetails.ProductID
    INNER JOIN(
      SELECT *, SUM(Quantity) AS TotalQuantity
      FROM OrderDetails
      GROUP BY ProductID
      ORDER BY TotalQuantity DESC
    LIMIT 1) AS TopSaleProduct
    ON TopSaleProduct.ProductID = OrderDetails.ProductID
    ''',
    con,
)
```

6. Выведите топ 5 поставщиков по количеству заказов, выведите их названия, страну, контактное лицо и телефон.

```python
tasks+=1
pd.read_sql(
    '''
    SELECT Country, SupplierName, ContactName, Phone FROM Suppliers INNER JOIN (
      SELECT SupplierID, COUNT(DISTINCT OrderID) AS OrderCnt FROM OrderDetails
      INNER JOIN Products ON OrderDetails.ProductID = Products.ProductID
      GROUP BY SupplierID ORDER BY OrderCnt DESC LIMIT 5) as TOP5_Supp
    ON Suppliers.SupplierID = TOP5_Supp.SupplierID

    ''',
    con,
)
```

7. Какую категорию товаров заказывали больше всего по стоимости в Бразилии, выведите страну, название категории и сумму.

```python
tasks+=1
pd.read_sql(
    '''
   SELECT Country, CategoryName, SUM(PositionPrice) AS PriceByCategory
    FROM(SELECT *, (Quantity * Price) AS PositionPrice FROM Products
        INNER JOIN OrderDetails ON Products.ProductID = OrderDetails.ProductID
        INNER JOIN Orders ON Orders.OrderID = OrderDetails.OrderID
        INNER JOIN Customers ON Orders.CustomerID = Customers.CustomerID
        INNER JOIN Categories ON Categories.CategoryID = Products.CategoryID
        WHERE Country = 'Brazil') AS Tbl
    GROUP BY CategoryID
    ORDER BY PriceByCategory DESC
    LIMIT 1
    ''',
    con,
)
```

8. Какая разница в стоимости между самым дорогим и самым дешевым заказом из США.

```python
tasks+=1
pd.read_sql(
    '''
    SELECT OrderID,MAX(OrderPrice) AS MaxOrderPrice, MIN(OrderPrice) AS MinOrderPrice, MAX(OrderPrice) - MIN(OrderPrice) AS PriceDiff
    FROM(SELECT OrderID, SUM(PositionPrice) AS OrderPrice
        FROM( SELECT *, (Quantity * Price) AS PositionPrice
            FROM Orders INNER JOIN Customers ON Orders.CustomerID = Customers.CustomerID
                        INNER JOIN OrderDetails ON OrderDetails.OrderID = Orders.OrderID
                        INNER JOIN Products ON OrderDetails.ProductID = Products.ProductID
            WHERE Country = 'USA') AS PositionPrice
    GROUP BY OrderID) AS MaxMinPrice
    ''',
    con,
)
```

9. Выведите количество заказов у каждого их трех самых молодых сотрудников, а также имя и фамилию во второй колонке.

```python
tasks+=1
pd.read_sql(
    '''
    SELECT COUNT(OrderID) AS OrdersCount, (FirstName ||' '|| LastName) AS FullName
    FROM Orders INNER JOIN (SELECT * FROM Employees
    ORDER BY substr(BirthDate, 7) || substr(BirthDate, 4, 2) || substr(BirthDate, 1, 2) DESC, EmployeeID ASC LIMIT 3) AS Top3YoungEmployee ON Top3YoungEmployee.EmployeeID = Orders.EmployeeID
    GROUP BY FullName
    ORDER BY OrdersCount ASC
    ''',
    con,
)
```
