3. Выведите колонки, у которых медиана равна минимальному значению (название колонки) выбирая из CRIM, ZN, INDUS, CHAS, NOX.
Напишите какой вывод можно сделать по данным в этих колонках

```python
tasks+=1
pd.read_sql(
    '''
    SELECT *
    FROM
    (
    SELECT 'CRIM' AS name, (SELECT MIN(CRIM) FROM boston) AS min,
    (SELECT AVG(CRIM) FROM (SELECT CRIM FROM boston ORDER BY CRIM
    LIMIT 2 - (SELECT COUNT(*) FROM boston) % 2    -- odd 1, even 2
    OFFSET (SELECT (COUNT(*) - 1) / 2 FROM boston))) AS median

    UNION
    SELECT 'ZN' AS name, (SELECT MIN(ZN) FROM boston) AS min,
    (SELECT AVG(ZN) FROM (SELECT ZN FROM boston ORDER BY ZN
    LIMIT 2 - (SELECT COUNT(*) FROM boston) % 2    -- odd 1, even 2
    OFFSET (SELECT (COUNT(*) - 1) / 2 FROM boston))) AS median

    UNION
    SELECT 'INDUS' AS name, (SELECT MIN(INDUS) FROM boston) AS min,
    (SELECT AVG(INDUS) FROM (SELECT INDUS FROM boston ORDER BY INDUS
    LIMIT 2 - (SELECT COUNT(*) FROM boston) % 2    -- odd 1, even 2
    OFFSET (SELECT (COUNT(*) - 1) / 2 FROM boston))) AS median

    UNION
    SELECT 'CHAS' AS name, (SELECT MIN(CHAS) FROM boston) AS min,
    (SELECT AVG(CHAS) FROM (SELECT CHAS FROM boston ORDER BY CHAS
    LIMIT 2 - (SELECT COUNT(*) FROM boston) % 2    -- odd 1, even 2
    OFFSET (SELECT (COUNT(*) - 1) / 2 FROM boston))) AS median

    UNION
    SELECT 'NOX' AS name, (SELECT MIN(NOX) FROM boston) AS min,
    (SELECT AVG(NOX) FROM (SELECT NOX FROM boston ORDER BY NOX
    LIMIT 2 - (SELECT COUNT(*) FROM boston) % 2    -- odd 1, even 2
    OFFSET (SELECT (COUNT(*) - 1) / 2 FROM boston))) AS median

    )
    WHERE (min - median) = 0
   ''',
  con,
)
```

```python
#Вывод:
#В столбцах CHAS и ZN больше всего нулевых значений, поэтому их медиана также ровна нулю, как и минимальное значение.
```

4. Выведите разницу между среднем количеством комнат(RM) в домах с самой дорогой стоимостью(MEDV) и 25 самыми дешевыми домами.
Аналогично по 50, 100, 200, 300 самыми дешевыми домами. (кол-во домов(25,50,100,200,300), среднее кол-во комнат в них, среднее кол-во комнат в самых дорогих, разница).
Напишите влияет ли кол-во комнат на стоимость и как сильно.

```python
tasks+=1
pd.read_sql(
    '''
    SELECT *,
    (avg_RM_by_max_MEDV - avg_RM_by_cheapest_homes) AS diff
    FROM(

    SELECT 25 AS cheapest_homes_count,
    (SELECT AVG(RM) FROM (SELECT * FROM boston ORDER BY MEDV LIMIT 25)) AS avg_RM_by_cheapest_homes,
    (SELECT AVG(RM) FROM boston WHERE MEDV = (SELECT MAX(MEDV) FROM boston)) AS avg_RM_by_max_MEDV

    UNION
    SELECT 50 AS cheapest_homes_count,
    (SELECT AVG(RM) FROM (SELECT * FROM boston ORDER BY MEDV LIMIT 50)) AS avg_RM_by_cheapest_homes,
    (SELECT AVG(RM) FROM boston WHERE MEDV = (SELECT MAX(MEDV) FROM boston)) AS avg_RM_by_max_MEDV

    UNION
    SELECT 100 AS cheapest_homes_count,
    (SELECT AVG(RM) FROM (SELECT * FROM boston ORDER BY MEDV LIMIT 100)) AS avg_RM_by_cheapest_homes,
    (SELECT AVG(RM) FROM boston WHERE MEDV = (SELECT MAX(MEDV) FROM boston)) AS avg_RM_by_max_MEDV

    UNION
    SELECT 200 AS cheapest_homes_count,
    (SELECT AVG(RM) FROM (SELECT * FROM boston ORDER BY MEDV LIMIT 200)) AS avg_RM_by_cheapest_homes,
    (SELECT AVG(RM) FROM boston WHERE MEDV = (SELECT MAX(MEDV) FROM boston)) AS avg_RM_by_max_MEDV

    UNION
    SELECT 300 AS cheapest_homes_count,
    (SELECT AVG(RM) FROM (SELECT * FROM boston ORDER BY MEDV LIMIT 300)) AS avg_RM_by_cheapest_homes,
    (SELECT AVG(RM) FROM boston WHERE MEDV = (SELECT MAX(MEDV) FROM boston)) AS avg_RM_by_max_MEDV

    )
    ''',
    con,
)
```

```python
#Выводы:
#Количество комнат, особенно с увеличением выборки (соответсвенно и с ее усреднением) не влияет на стоимость домов. Вероятно, на это влияют другие параметры, например, расположение дома.
```
