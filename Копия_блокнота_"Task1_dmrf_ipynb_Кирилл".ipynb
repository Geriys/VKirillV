## Шаг 1й - Получение списка объектов

```python
# Для скачивания данных

offset_ = 1 # размер шага
limit_ = 1000 # количество объектов в выдаче
url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/kn/object?offset={offset_}&limit={limit_}&sortField=devId.devShortCleanNm&sortType=asc&objStatus=0'
res = requests.get(url)
objects_data = res.json()
objects_data.get('data').get('list')[0] # К сожалению не понимаю, что такое list, подставляю лимит_ пишет none (это в следующей ячейке)
```

```python
objects_data.keys()
```

```python
objects_list = objects_data.get('data').get('list')
objids = [x.get('objId') for x in objects_list]
len(objids)
#objids1 = [x1.get('siteStatus') for x1 in objects_list]
objids[:1000]
```

```python
len(objids)
```

```python
full_obj = []
for dom_id in tqdm(objids):
    url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/object/{dom_id}'
    ress = requests.get(url)
    one_dom = ress.json()
    full_obj.append(one_dom)
```

```python
# Создаем таблицу DataFrame из всех полученных данных
import pandas as pd
df = pd.json_normalize(full_obj)
```
